```python
import numpy as np
import pandas as pd 
```

```python
from mplsoccer import Pitch, VerticalPitch
```

```python
df = pd.read_csv('./erling_haaland_2022_understat.csv')

df.head(5)
```

```python
df.X = df.X * 100
df.Y = df.Y * 100

df.head(3)
```

```python
total_shots = df.shape[0]
total_goals = df [df['result']  == 'Goal'].shape[0]
Xg = df.xG.sum()     
xG_per_Shot = Xg / total_shots  
avg_distance = df.X.mean()
actual_avg_distance = 115 - (df.X * 1.15).mean()
```

```python
shot_counts = df['shotType'].value_counts()


droit  = shot_counts[0]

gauche  = shot_counts[1]


tete  = shot_counts[2]

```

```python
player = df.player[0]
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
font_path = './Arvo-Regular.ttf'
font_props = font_manager.FontProperties(fname=font_path)

```

```python
fig = plt.figure(figsize=(8,12))

background_color = '#0C0D0E'

fig.patch.set_facecolor(background_color)

ax1 = fig.add_axes([0, 0.7, 1, .2])
ax1.set_facecolor(background_color)
ax1.set_xlim(0, 1)
ax1.set_ylim(0, 1)

#First section 
ax1.text(
    x=.5,
    y=.85,
    s= f'{player}',
    fontsize = 20,
    fontproperties = font_props,
    fontweight = 'bold',
    color = 'white',
    ha='center'
)
#Titre
ax1.text(
    x=.5,
    y=.75,
    s= 'Tout les tire en Premmier League 2022-2023',
    fontsize = 13,
    fontproperties = font_props,
    fontweight = 'bold',
    color = 'white',
    ha='center'
)

#Attaque moins dang
ax1.text(
    x=.23,
    y=.5,
    s= 'Attaque moins dangereux',
    fontsize = 11,
    fontproperties = font_props,
    fontweight = 'bold',
    color = 'white',
    ha='center'
)

#Pts de l'echelle dang
ax1.scatter(
    x = .37,
    y = .515, 
    s = 100,
    color= background_color,
    edgecolor = 'white',
    linewidth = .8
)

#Pts de l'echelle dang
ax1.scatter(
    x = .42,
    y = .515, 
    s = 200,
    color= background_color,
    edgecolor = 'white',
    linewidth = .8
)


#Pts de l'echelle dang
ax1.scatter(
    x = .48,
    y = .515, 
    s = 300,
    color= background_color,
    edgecolor = 'white',
    linewidth = .8
)



#Pts de l'echelle dang
ax1.scatter(
    x = .54,
    y = .515, 
    s = 400,
    color= background_color,
    edgecolor = 'white',
    linewidth = .8
)



#Pts de l'echelle dang
ax1.scatter(
    x = .6,
    y = .515, 
    s = 500,
    color= background_color,
    edgecolor = 'white',
    linewidth = .8
)


#Attaque plus dang
ax1.text(
    x=.75,
    y=.5,
    s= 'Attaque plus dangereux',
    fontsize = 11,
    fontproperties = font_props,
    fontweight = 'bold',
    color = 'white',
    ha='center'
)


#Pt rouge pour but
ax1.scatter(
    x = .41,
    y = .29, 
    s = 100,
    color= 'red',
    edgecolor = 'white',
    linewidth = .8
)

#But
ax1.text(
    x=.45,
    y=.27,
    s= 'But',
    fontsize = 11,
    fontproperties = font_props,
    fontweight = 'bold',
    color = 'white',
    ha='center'
)

#Pt rnoire pour raté
ax1.scatter(
    x = .51,
    y = .29, 
    s = 100,
    color= background_color,
    edgecolor = 'white',
    linewidth = .8
)

#Raté
ax1.text(
    x=.55,
    y=.27,
    s= 'Raté',
    fontsize = 11,
    fontproperties = font_props,
    fontweight = 'bold',
    color = 'white',
    ha='center'
)


#Section 2
ax2 = fig.add_axes([.05, 0.25, .9, .5])
ax2.set_facecolor(background_color)


#Terrain

pitch= VerticalPitch(
    pitch_type='opta',
    half= True,
    pitch_color='#136d15',
    pad_bottom=.5,
    line_color='white',
    linewidth=.75,
    axis= True,
    label= True
)

pitch.draw(ax=ax2)

#Distance Moy
ax2.scatter(x=90,y=avg_distance,s=100,color='white',linewidths=.8)
ax2.plot([90,90],[100,avg_distance],color = 'white' , linewidth=2)

ax2.text(
    x = 90,
    y= avg_distance - 7,
    s = f'Distance\n Moyenne\n {actual_avg_distance :.1f} yards' ,
    color = 'white',
    fontsize = 11.5,
    fontproperties = font_props,
    ha='center',
    fontweight = 'bold',
)







for x in df.to_dict(orient='records'):
    pitch.scatter(
        x['X'],
        x['Y'],
        s=300 * x['xG'],
        color = 'red' if x['result'] == 'Goal' else background_color,
        edgecolor='white',
        alpha=.7,
        linewidth=.8,
        ax=ax2
    )



#Tires Totale
ax2.text(
    x = -18,
    y= 86,
      s = f'Tires Totale' ,
    color = 'white',
     fontsize = 20,
    fontproperties = font_props,
    ha='center',
    fontweight='bold',
)


ax2.text(
    x = -18,
    y= 83,
    s = f'{total_shots}' ,
    color = 'red',
     fontsize = 16,
    fontproperties = font_props,
    ha='center',
)



#Droit
ax2.text(
    x = -18,
    y= 76,
      s = f'Pied Droit' ,
    color = 'white',
     fontsize = 20,
    fontproperties = font_props,
    ha='center',
    fontweight='bold',
)

ax2.text(
    x = -18,
    y= 73,
    s = f'{droit}' ,
    color = 'red',
     fontsize = 16,
    fontproperties = font_props,
    ha='center',
)






#Gauche
ax2.text(
    x = -18,
    y= 66,
      s = f'Pied Gauche' ,
    color = 'white',
     fontsize = 20,
    fontproperties = font_props,
    ha='center',
    fontweight='bold',
)


ax2.text(
    x = -18,
    y= 63,
     s = f'{gauche}' ,
    color = 'red',
     fontsize = 16,
    fontproperties = font_props,
    ha='center',
)





#tete
ax2.text(
    x = -18,
    y= 56,
      s = f'Tete' ,
    color = 'white',
     fontsize = 20,
    fontproperties = font_props,
    ha='center',
    fontweight='bold',
)


ax2.text(
    x = -18,
    y= 53,
     s = f'{tete}' ,
    color = 'red',
     fontsize = 16,
    fontproperties = font_props,
    ha='center',
)




#Section 3
ax3 = fig.add_axes([0, .2, 1, .05])
ax3.set_facecolor(background_color)




#Totale Buts
ax3.text(
  x = .2,
    y= .5,
    s = f'Goals' ,
    color = 'white',
     fontsize = 20,
    fontproperties = font_props,
    ha='center',
    fontweight='bold',
)

ax3.text(
   x = .2,
    y= .0,
    s = f'{total_goals}' ,
    color = 'red',
     fontsize = 15,
    fontproperties = font_props,
    ha='center',
)





#Totale XG
ax3.text(
    x = .5,
    y= .5,
    s = f'xG' ,
    color = 'white',
     fontsize = 20,
    fontproperties = font_props,
    ha='center',
    fontweight='bold',
)

ax3.text(
    x = .5,
    y= .0,
    s = f'{Xg:.1f}' ,
    color = 'red',
     fontsize = 15,
    fontproperties = font_props,
    ha='center',
)


#xG par Tire

ax3.text(
    x = .8,
    y= .5,
    s = f'xG/Tire' ,
    color = 'white',
     fontsize = 20,
    fontproperties = font_props,
    ha='center',
    fontweight='bold',
)

ax3.text(
    x = .8,
    y= .0,
    s = f'{xG_per_Shot:.2f}' ,
    color = 'red',
     fontsize = 15,
    fontproperties = font_props,
    ha='center',
)




```
